```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. Tách đặc trưng (X) và nhãn (y)
X = df.drop('Class', axis=1)
y = df['Class']

# 2. Mã hóa nhãn (Chuyển từ chữ sang số)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print("Các loại đậu đã được mã hóa thành:", list(le.classes_))

# 3. Chia tập Train/Test (80% Train, 20% Test)
# stratify=y đảm bảo tỷ lệ các loại đậu trong tập Test giống tập Train
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# 4. Chuẩn hóa dữ liệu (Standard Scaling)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nKích thước tập huấn luyện: {X_train_scaled.shape}")
print(f"Kích thước tập kiểm thử: {X_test_scaled.shape}")
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Kiểm tra phân bố trước khi SMOTE
print("Trước khi SMOTE:", Counter(y_train))

# Áp dụng SMOTE chỉ trên tập TRAIN (Không được chạm vào tập Test để tránh rò rỉ dữ liệu)
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

# Kiểm tra phân bố sau khi SMOTE
print("Sau khi SMOTE:  ", Counter(y_train_balanced))
print("✅ Dữ liệu đã được cân bằng hoàn hảo!")
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import time

# Hàm huấn luyện và đánh giá chung
def train_evaluate_model(model, name, X_train, y_train, X_test, y_test):
    print(f"\n--- Bắt đầu huấn luyện: {name} ---")
    start_time = time.time()

    # Huấn luyện
    model.fit(X_train, y_train)

    # Dự đoán
    y_pred = model.predict(X_test)

    # Đánh giá
    end_time = time.time()
    accuracy = accuracy_score(y_test, y_pred)
    print(f"⏱️ Thời gian huấn luyện: {end_time - start_time:.2f} giây")
    print(f"🎯 Độ chính xác (Accuracy): {accuracy:.4f}")

    return y_pred, accuracy

# Khởi tạo 3 mô hình
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42),
    "Neural Network (MLP)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
}

# Chạy vòng lặp huấn luyện
results = {}
predictions = {}

for name, model in models.items():
    # Sử dụng dữ liệu đã cân bằng (X_train_balanced) để huấn luyện
    y_pred, acc = train_evaluate_model(model, name, X_train_balanced, y_train_balanced, X_test_scaled, y_test)
    results[name] = acc
    predictions[name] = y_pred
```

```python
# 1. Biểu đồ so sánh độ chính xác
plt.figure(figsize=(10, 6))
sns.barplot(x=list(results.keys()), y=list(results.values()), palette='magma')
plt.title('So sánh Độ chính xác giữa các Mô hình')
plt.ylim(0.8, 1.0) # Giới hạn trục Y để nhìn rõ sự chênh lệch
plt.ylabel('Accuracy')
for i, v in enumerate(results.values()):
    plt.text(i, v + 0.005, f"{v:.4f}", ha='center', fontweight='bold')
plt.show()

# 2. Phân tích chi tiết mô hình tốt nhất (Ví dụ lấy Random Forest hoặc model có acc cao nhất)
best_model_name = max(results, key=results.get)
print(f"\n🏆 Mô hình tốt nhất là: {best_model_name}")

# In báo cáo chi tiết
print("\n--- Classification Report (Chi tiết từng loại đậu) ---")
print(classification_report(y_test, predictions[best_model_name], target_names=le.classes_))

# 3. Vẽ Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_test, predictions[best_model_name])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title(f'Ma trận nhầm lẫn (Confusion Matrix) - {best_model_name}')
plt.xlabel('Dự đoán (Predicted)')
plt.ylabel('Thực tế (Actual)')
plt.show()
```

```python
# --- CELL 13: REAL-TIME SYSTEM SIMULATION (DEMO) ---
import random

def simulate_real_time_classification(n_samples=5):
    print(f"--- 📸 KHỞI ĐỘNG HỆ THỐNG CAMERA PHÂN LOẠI ({n_samples} mẫu thử) ---\n")
    print(f"{'STT':<5} | {'Diện tích':<10} | {'Chu vi':<10} | {'Dự đoán AI':<15} | {'Thực tế':<15} | {'Kết quả':<10}")
    print("-" * 80)

    # Lấy ngẫu nhiên n mẫu từ tập Test
    indices = random.sample(range(len(X_test_scaled)), n_samples)

    correct_count = 0

    for i, idx in enumerate(indices):
        # Lấy dữ liệu 1 hạt đậu (giả lập camera vừa chụp xong và xử lý)
        sample = X_test_scaled[idx].reshape(1, -1)
        actual_label_idx = y_test[idx]
        actual_label_str = le.inverse_transform([actual_label_idx])[0]

        # Lấy thông số gốc (để hiển thị cho sinh động)
        # Lưu ý: X_test gốc nằm ở biến X_test (trước khi scale), ta cần lấy lại để in ra số thực
        original_area = X_test.iloc[idx]['Area']
        original_peri = X_test.iloc[idx]['Perimeter']

        # AI Dự đoán
        pred_label_idx = best_model.predict(sample)[0]
        pred_label_str = le.inverse_transform([pred_label_idx])[0]

        # Kiểm tra đúng sai
        is_correct = "✅ ĐÚNG" if pred_label_idx == actual_label_idx else "❌ SAI"
        if pred_label_idx == actual_label_idx: correct_count += 1

        print(f"{i+1:<5} | {original_area:<10.0f} | {original_peri:<10.2f} | {pred_label_str:<15} | {actual_label_str:<15} | {is_correct}")

    print("-" * 80)
    print(f"kết quả phiên làm việc: Chính xác {correct_count}/{n_samples} mẫu.")

# Chạy mô phỏng
simulate_real_time_classification(10)
```

```python
# 1. Vẽ Boxplot cho đặc trưng Diện tích (Area)
plt.figure(figsize=(14, 8))
sns.boxplot(x='Class', y='Area', data=df, palette='Set2')
plt.title('Phân tích ngoại lệ: Diện tích (Area) theo từng loại đậu')
plt.xlabel('Loại đậu')
plt.ylabel('Diện tích (pixels)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# 2. Đếm số lượng ngoại lệ bằng thuật toán IQR (Interquartile Range)
def count_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    # Công thức chuẩn xác định ngoại lệ: nằm ngoài khoảng [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
    outliers = ((series < (Q1 - 1.5 * IQR)) | (series > (Q3 + 1.5 * IQR))).sum()
    return outliers

print("📊 Số lượng điểm ngoại lệ (Outliers) trên đặc trưng Diện tích:")
outlier_counts = df.groupby('Class')['Area'].apply(count_outliers_iqr)
print(outlier_counts)
```

```python
# --- YÊU CẦU 3: ĐÁNH GIÁ ĐẶC TRƯNG (CORRELATION & FEATURE EVALUATION) ---

# 1. Tính ma trận tương quan (chỉ lấy cột số)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numerical_cols].corr()

# 2. Vẽ Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Ma trận tương quan giữa các đặc trưng (Feature Correlation)')
plt.show()

# 3. Lọc ra các cặp đặc trưng tương quan rất mạnh (> 0.95) để cảnh báo
print("👉 CÁC CẶP ĐẶC TRƯNG CÓ TƯƠNG QUAN RẤT MẠNH (> 0.95):")
# Lấy tam giác trên của ma trận để không in trùng lặp
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_pairs = [(column, row) for row in upper_tri.index for column in upper_tri.columns
                   if abs(upper_tri.loc[row, column]) > 0.95]

for pair in high_corr_pairs:
    print(f"- {pair[0]} và {pair[1]}: {corr_matrix.loc[pair[0], pair[1]]:.4f}")
print("(Nhận xét: Các đặc trưng này mang thông tin dư thừa giống nhau)")
```
